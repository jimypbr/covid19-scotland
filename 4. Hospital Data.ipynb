# Hospital Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('darkgrid')
```

```python
df_hos = pd.read_pickle('./corona_scotland_regions_hospital_ts.pk')
df_icu = pd.read_pickle('./corona_scotland_regions_icu_ts.pk')
```

```python
df_hos = df_hos.reset_index(drop=True)
df_icu = df_icu.reset_index(drop=True)
```

## Scotland (Total)

```python
df_hos['Scotland (Total)'].plot.bar()
```

```python
df_icu['Scotland (Total)'].plot.bar()
```

## Regions

```python
axes = df_hos.plot.bar(subplots=True, sharey=False, figsize=(20,20), layout=(6, 3))
for ax in axes.flatten(): 
    ax.set_ylabel('Num In Hospital')
    ax.set_xlabel('Days since 08/04')
plt.tight_layout()
plt.show()
```

```python
axes = df_icu.plot.bar(subplots=True, sharey=False, figsize=(20,20), layout=(6, 3))
for ax in axes.flatten(): 
    ax.set_ylabel('Num In ICU')
    ax.set_xlabel('Days since 08/04')
plt.tight_layout()
plt.show()
```
