# Hospital Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('darkgrid')
```

```python
df_hos = pd.read_pickle('./corona_scotland_regions_hospital_ts.pk')
df_icu = pd.read_pickle('./corona_scotland_regions_icu_ts.pk')
```

## Scotland (Total)

```python
df_hos['Scotland (Total)'].plot.bar()
```

```python
df_icu['Scotland (Total)'].plot.bar()
```

## Regions

```python
axes = df_hos.drop(columns='Scotland (Total)').plot.bar(subplots=True, sharey=True, figsize=(20,20), layout=(4, 4))
for ax in axes.flatten(): 
    ax.set_ylabel('Num In Hospital')
    ax.set_xlabel('Date')
plt.tight_layout()
plt.show()
```
